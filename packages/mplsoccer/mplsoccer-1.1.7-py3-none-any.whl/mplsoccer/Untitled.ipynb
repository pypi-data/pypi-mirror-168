```python
def inset_axes(x, y, width=None, height=None, aspect=None, ax=None, vertical=False, **kwargs):
    if vertical:
        x, y = y, x
    ax_aspect = ax.get_aspect()
    if width is None and aspect:
        width = height * ax_aspect * aspect
    x = x - width / 2
    y = y - height / 2
    return ax.inset_axes((x, y, width, height), transform=ax.transData, **kwargs)
```
